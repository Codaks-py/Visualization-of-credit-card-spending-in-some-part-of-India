```python
import pandas as pd
import numpy as np
import matplotlib as plt
import seaborn as sns

%matplotlib inline
```

```python
df = df1.drop_duplicates('Date')
df
```

```python
del df['index']
```

<b><i>Females spends more

```python
df.Gender.value_counts().plot(kind='bar')
```

```python
df=df.rename(
    columns={
        'Exp Type' : 'Exp_Type',
        'Card Type' : 'Card_Type'
    }
)
```

<b>Greater Mumbai, India records more spending than other cities

```python
df.City.value_counts().plot(kind='barh', figsize=(10,5), title='List of City in India with highest Transaction'
)
```

<b><i> We can deduce by city what individuals spend spend their money on using different card,
<i>And also people spend more on bills

```python
df[
    df.City.isin([
        'Greater Mumbai, India', 'Delhi, India', 'Bengaluru, India', 'Ahmedabad, India'
    ])
].groupby(
    ['City', 'Card_Type']
).Exp_Type.value_counts().unstack().plot.bar(
    title =' Cities with  Card Types For Transactions and how they spend their money', figsize = (20, 5)
)

```

<i><b>We can dedudes that Females dominates in all card type and cities

```python
df[
    df.City.isin([
        'Greater Mumbai, India', 'Delhi, India', 'Bengaluru, India', 'Ahmedabad, India'
    ])
].groupby(
    ['City', 'Card_Type']
).Gender.value_counts().unstack().plot.bar(
    title =' Genders with Highest Card Types For Transactions Per Cities', figsize = (20, 5)
)
```

<i><b> We can Agree That People hold more of Silver Card than the rest in these cities

```python
df[
    df.City.isin([
        'Greater Mumbai, India', 'Delhi, India', 'Bengaluru, India', 'Ahmedabad, India'
    ])
].groupby(
    ['City', 'Card_Type']
).Card_Type.value_counts().unstack().plot.bar(
    title =' Showing The level of Card type in each City', figsize = (20, 5)
)
```

<i><b>Showing the total in figures spent by each genders
